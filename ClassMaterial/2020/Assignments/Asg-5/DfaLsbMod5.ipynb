```python
def test_dfalsb(dfa, n):
    all_pass = True
    for i in range(1, n):
        s = nthnumeric(i, Sigma={'0','1'})
        real_val = int(s[::-1], 2)
        should_pass = real_val % 5 == 0
        if should_pass != accepts_dfa(dfa, s):
            all_pass = False
            print('Test failed:', s, real_val, real_val%5)
    if all_pass:
        print('All Tests Passed!')
```
