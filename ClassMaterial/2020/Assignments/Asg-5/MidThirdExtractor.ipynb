```python
def MidThirdExtractor(D):
    ### We first create an NFA since they are easier to build and then reduce to a minimal DFA
    D_q, D_sig, D_delta, D_q0, D_f = D['Q'], D['Sigma'], D['Delta'], D['q0'], D['F']
    N_q     = set(product(D_q, D_q, D_q, D_q, D_q))                           # Q^5 of initial DFA
    N_sig   = D_sig                                                           # Keeps the same alphabet
    N_q0    = set({(D_q0, q1, q1, q2, q2) for q1 in D_q for q2 in D_q})
    N_F     = set({(q1, q1, q2, q2, qf) for q1 in D_q for q2 in D_q for qf in D_q})
    N_delta = {}
    # Writing the delta function with loops since one line comprehension is too messy
    for qa, q1, qb, q2, qc in N_q:
        for y in N_sig:
            N_delta[((qa, q1, qb, q2, qc), y)] = set({(D_delta[(qa, c1)], q1, D_delta[(qb, y)], q2, D_delta[(qc, c2)])
                                                         for c1 in N_sig for c2 in N_sig})
    N = mk_nfa(N_q, N_sig, N_delta, N_q0, N_F)
    D_middle_third = nfa2dfa(N)
    return min_dfa(D_middle_third)
```

```python
def test_Mid_Third_Extractor(re_first, re_middle, re_last, test_set):
    re_total = re_first + re_middle + re_last
    D_total = nfa2dfa(re2nfa(re_total))
    D_total = min_dfa(D_total)
    D_mid = MidThirdExtractor(D_total)
    pass_all = True
    for s in test_set:
        if accepts_dfa(D_mid, s) == False:
            print('DFA Failed to accept', s)
            pass_all = False
    if pass_all:
        print('Test worked! RE:', re_total)
    else:
        print('Test failed. RE:', re_total)
```
